### User defined functions

> DataFrame.agg() and DataFrame.transform() applies a user defined function that reduces or broadcasts its result respectively.

Aggregate
> The result is aggregated or summarized. You get a single value (like mean, sum, or count) for each column.

```python
pkfile.select_dtypes(include='number').agg(lambda x: x.mean() * 2)
```

Transform
> The result is broadcasted to match the shape of the input. You get the same number of rows and columns as the original DataFrame.

```python
pkfile.select_dtypes(include= "number").transform(lambda x: x * 10)
```

### Grouping

By “group by” we are referring to a process involving one or more of the following steps:

-   Splitting the data into groups based on some criteria

-   Applying a function to each group independently

-   Combining the results into a data structure

```python
#Creating Data First

gdata = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

gdata
```

##### Grouping by a column label, selecting column labels, and then applying the DataFrameGroupBy.sum() function to the resulting groups:

```python
gdata.groupby("A")[["C","D"]].sum()
```

##### Grouping by multiple columns label forms MultiIndex.

```python
gdata.groupby(["A", "B"]).sum()
```

### Stack

> In Pandas, the stack() function is used to pivot the columns of a DataFrame into rows, essentially converting a wide format DataFrame into a long format.

```python
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)
print(df)
```

```python
stacked_df = df.stack()
print(stacked_df)
```

```python
unstacked_df = stacked_df.unstack()
print(unstacked_df)
```

### Pivot Table

> In Pandas, a pivot table is a powerful tool that allows you to summarize and reorganize data by applying aggregation functions (like sum, mean, etc.) on columns based on specific categories or index levels.

```python
data = {
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Region': ['North', 'North', 'South', 'South', 'North', 'South'],
    'Sales': [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)
print(df)
```

```python
pivot_table = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot_table)
```

Adding Margins (Subtotals):


```python
pivot_table_margins = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc='sum', margins=True)
print(pivot_table_margins)
```

##### Applying Multiple Aggregation Functions:


```python
pivot_table_multiagg = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc=['sum', 'mean'])
print(pivot_table_multiagg)
```

## **Time series**

1. Datetime Conversion:


```python
# Sample data
data = {'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04'],
        'Sales': [200, 220, 250, 300]}

df = pd.DataFrame(data)

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

print(df)
```

2. Setting a Datetime Index:


```python
df.set_index('Date', inplace=True)
print(df)
```

3. Resampling:


```python
# Resample to sum sales over each month
monthly_sales = df.resample('ME').sum()
print(monthly_sales)
```

### **Categoricals**

```python
df = pd.DataFrame(
        {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
   )
df
```

> Converting the raw grades to a categorical data type:



```python
df["grade"] = df["raw_grade"].astype("category")
df["grade"]
```

> Rename the categories to more meaningful names:



```python
new_categories = ["very good", "good", "very bad"]
df["grade"] = df["grade"].cat.rename_categories(new_categories)
df["grade"]

```

> Reorder the categories and simultaneously add the missing categories

```python
df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"]
)
df["grade"]
```

> Sorting is per order in the categories, not lexical order:



```python
df.sort_values(by="grade")
```

> Grouping by a categorical column with observed=False also shows empty categories:



```python
df.groupby("grade", observed=False).size()
```
