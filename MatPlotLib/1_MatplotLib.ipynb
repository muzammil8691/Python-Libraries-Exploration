```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **A simple example**


```python
xpoints = np.array([1,5])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints)
plt.show()
```

### **Plotting Without Line**


```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

### **Multiple Points**

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

#Line Graph
plt.plot(xpoints, ypoints)
plt.show()

#PointGraph
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

### **Default X-Points**

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

# **Markers**
You can use the keyword argument marker to emphasize each point with a specified marker:

### Marker Reference

> You can choose any of these markers:

Marker	Description

'o'	Circle	

'*'	Star	

'.'	Point	

','	Pixel	

'x'	X	

'X'	X (filled)	

'+'	Plus	

'P'	Plus (filled)	

's'	Square	

'D'	Diamond	

'd'	Diamond (thin)	

'p'	Pentagon	

'H'	Hexagon	

'h'	Hexagon	

'v'	Triangle Down	

'^'	Triangle Up	

'<'	Triangle Left	

'>'	Triangle Right	

'1'	Tri Down	

'2'	Tri Up	

'3'	Tri Left	

'4'	Tri Right	

'|'	Vline	

'_'	Hline

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

Mark each point with a star:



```python
plt.plot(ypoints, marker = '*')
```

### **Line Reference**

Line Syntax	Description

'-'	Solid line	

':'	Dotted line	

'--'	Dashed line	

'-.'	Dashed/dotted line

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o--')
plt.show()
```

**If you leave out the line value in the fmt parameter, no line will be plotted.**

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'x')
plt.show()
```

# **Color Reference**

Syntax	------ Description

'r'	------------ Red	

'g'	------------ Green	

'b'	------------ Blue	

'c'	------------ Cyan	

'm'	------------ Magenta	

'y'	------------ Yellow	

'k'	------------ Black	

'w'	------------ White

```python
ypoints = np.array([3, 8, 1, 10])


plt.plot(ypoints, 'x:g')
plt.show()
```

## Size of Marker
Set the size of the markers to 20:


```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'D', ms = 20)
plt.show()
```

### **Marker Color**

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:



```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'D', ms = 15, mec = 'r')
plt.show()
```

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

## **Linestyle**

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
plt.show()


# Line Syntax	Description

# '-'	Solid line	

# ':'	Dotted line	

# '--'	Dashed line	

# '-.'	Dashed/dotted line
```

#### Shorter Syntax


```python
plt.plot(ypoints, ls = '-.')
```

## **Line Color**

You can use the keyword argument color or the shorter c to set the color of the line:



```python
ypoints = np.array([3, 8, 1, 10])

# plt.plot(ypoints, color = 'r')
#or
plt.plot(ypoints, c = 'hotpink')
plt.show()

```

## **Line Width**

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20')
plt.show()
```

## **Multiple Lines**

You can plot as many lines as you like by simply adding more plt.plot() functions:



```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

## **Matplotlib Labels and Title**


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")              #Title
plt.xlabel("Average Pulse")                 #X Label
plt.ylabel("Calorie Burnage")               #Y Label

plt.show()
```

### **Matplotlib Adding Grid Lines**


```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

Specify Which Grid Lines to Display


```python
plt.plot(x, y)

plt.grid(axis = 'x') #Could be X or Y

plt.show()
```

### **Set Line Properties for the Grid**


```python
plt.plot(x, y)

plt.grid(color = 'red', linestyle = '--', linewidth = 2)

plt.show()
```

## **Matplotlib Subplot**


##### Display Multiple Plots


```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

#### **The subplot() Function**




```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) #the figure has 2 rows, 1 column, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2) #the figure has 2 rows, 1 column, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

#### Title

You can add a title to each plot with the title() function:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

#### Super Title
You can add a title to the entire figure with the suptitle() function:

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
